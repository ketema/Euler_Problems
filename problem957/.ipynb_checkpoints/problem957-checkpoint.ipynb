```python
# Cell: Line intersection and day transition functions
def line_intersection(p1, p2, p3, p4):
    """
    Find intersection of line (p1, p2) with line (p3, p4)
    Returns None if parallel
    """
    d1 = p2 - p1
    d2 = p4 - p3

    # Solve using determinant
    det = d1[0]*(-d2[1]) - d1[1]*(-d2[0])

    if det == 0:
        return None  # Parallel

    rhs = p3 - p1
    t = (rhs[0]*(-d2[1]) - rhs[1]*(-d2[0])) / det

    return p1 + t*d1

def compute_next_day(reds, blues):
    """
    Given current blues, compute all new intersection points
    """
    existing = set([tuple(r) for r in reds] + [tuple(b) for b in blues])

    # All lines through (red, blue) pairs
    lines = []
    for r in reds:
        for b in blues:
            lines.append((r, b))

    # Find all intersections
    new_points = set()
    for i, (p1, p2) in enumerate(lines):
        for (p3, p4) in lines[i+1:]:
            pt = line_intersection(p1, p2, p3, p4)
            if pt is not None:
                pt_tuple = tuple(pt)
                if pt_tuple not in existing and pt_tuple not in new_points:
                    new_points.add(pt_tuple)

    return [vector(QQ, pt) for pt in new_points]
```

```python
# Cell: Test the OPTIMAL parabola configuration
R1 = vector(QQ, [0, 0])
R2 = vector(QQ, [1, 0])
R3 = vector(QQ, [0, 1])

# Blues on parabola y = x² (OPTIMAL positions)
B1 = vector(QQ, [1, 1])    # (1, 1²)
B2 = vector(QQ, [3, 9])    # (3, 3²)

print("Red triangle:")
print(f"  R1 = {R1}")
print(f"  R2 = {R2}")
print(f"  R3 = {R3}")
print("\nBlues on parabola y=x² (OPTIMAL):")
print(f"  B1 = {B1}")
print(f"  B2 = {B2}")

# Test it
reds = [R1, R2, R3]
blues = [B1, B2]

new_blues_day1 = compute_next_day(reds, blues)
print(f"\nDay 0 → 1: Generated {len(new_blues_day1)} new blues")
print(f"Total g(1) = {len(blues) + len(new_blues_day1)}")

# Show the new points
print("\nNew blue points created:")
for i, pt in enumerate(new_blues_day1, 1):
    print(f"  New_{i} = {pt}")
```

```python
# Cell: Visualize the configuration
import matplotlib.pyplot as plt

def plot_day_zoomed(reds, blues, new_blues, title="Point Genesis"):
    """Plot the configuration with better zoom and smaller markers"""
    fig, ax = plt.subplots(figsize=(12, 12))
    
    # Plot red points (smaller squares)
    red_x = [float(r[0]) for r in reds]
    red_y = [float(r[1]) for r in reds]
    ax.scatter(red_x, red_y, c='red', s=120, marker='s', 
               label='Red (fixed)', zorder=5, edgecolors='black', linewidths=2)
    
    # Plot initial blue points (smaller size)
    blue_x = [float(b[0]) for b in blues]
    blue_y = [float(b[1]) for b in blues]
    ax.scatter(blue_x, blue_y, c='blue', s=70, marker='o', 
               label='Initial blues', zorder=4, edgecolors='black', linewidths=2)
    
    # Plot new blue points (larger and more prominent)
    if new_blues:
        new_x = [float(n[0]) for n in new_blues]
        new_y = [float(n[1]) for n in new_blues]
        ax.scatter(new_x, new_y, c='lime', s=130, marker='^', 
                   label='New blues', zorder=3, edgecolors='darkgreen', linewidths=2.5)
    
    # Draw EXISTING lines (red to existing blues) in orange
    for r in reds:
        for b in blues:
            dx = float(b[0] - r[0])
            dy = float(b[1] - r[1])
            if abs(dx) > 0.001 or abs(dy) > 0.001:
                t_extend = 3
                x1 = float(r[0]) - t_extend * dx
                x2 = float(r[0]) + t_extend * dx
                y1 = float(r[1]) - t_extend * dy
                y2 = float(r[1]) + t_extend * dy
                ax.plot([x1, x2], [y1, y2], 'orange', alpha=0.4, linewidth=1.2)
    
    # Draw NEW lines (red to new blues) in purple/magenta
    if new_blues:
        for r in reds:
            for n in new_blues:
                dx = float(n[0] - r[0])
                dy = float(n[1] - r[1])
                if abs(dx) > 0.001 or abs(dy) > 0.001:
                    t_extend = 3
                    x1 = float(r[0]) - t_extend * dx
                    x2 = float(r[0]) + t_extend * dx
                    y1 = float(r[1]) - t_extend * dy
                    y2 = float(r[1]) + t_extend * dy
                    ax.plot([x1, x2], [y1, y2], 'magenta', alpha=0.7, linewidth=2)
    
    # Label points
    for i, r in enumerate(reds, 1):
        ax.annotate(f'R{i}', (float(r[0]), float(r[1])), xytext=(8, 8), 
                    textcoords='offset points', fontsize=13, fontweight='bold', color='darkred')
    for i, b in enumerate(blues, 1):
        ax.annotate(f'B{i}', (float(b[0]), float(b[1])), xytext=(8, 8), 
                    textcoords='offset points', fontsize=12, color='darkblue', fontweight='bold')
    for i, n in enumerate(new_blues, 1):
        ax.annotate(f'N{i}', (float(n[0]), float(n[1])), xytext=(5, -12), 
                    textcoords='offset points', fontsize=10, color='darkgreen')
    
    # Zoom to region of interest
    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 5)
    
    ax.set_xlabel('x', fontsize=13)
    ax.set_ylabel('y', fontsize=13)
    ax.set_title(title, fontsize=15, fontweight='bold')
    ax.legend(fontsize=12, loc='upper left')
    ax.grid(True, alpha=0.4, linestyle='--')
    ax.axhline(y=0, color='k', linewidth=1, alpha=0.5)
    ax.axvline(x=0, color='k', linewidth=1, alpha=0.5)
    ax.set_aspect('equal')
    
    plt.tight_layout()
    plt.show()

# Visualize day 1 with better zoom
plot_day_zoomed(reds, blues, new_blues_day1, title="Day 0 → 1: g(1) = 8")
```

```python
# Cell: Search for optimal parabola configuration
print("Searching for optimal blue positions on parabola y=x²...")
print("Testing different x-coordinates for the two blues...\n")

best_config = None
best_g2 = 0

# Try different x-coordinates on the parabola
for x1 in range(1, 5):
    for x2 in range(x1+1, 6):
        # Blues on parabola y = x²
        test_blues = [vector(QQ, [x1, x1**2]), vector(QQ, [x2, x2**2])]
        
        # Compute g(1)
        new_day1 = compute_next_day(reds, test_blues)
        g1 = len(test_blues) + len(new_day1)
        
        if g1 == 8:  # Must get g(1) = 8
            # Compute g(2)
            all_day1 = test_blues + new_day1
            new_day2 = compute_next_day(reds, all_day1)
            g2 = len(all_day1) + len(new_day2)
            
            print(f"B1=({x1},{x1**2}), B2=({x2},{x2**2}): g(1)={g1}, g(2)={g2}")
            
            if g2 > best_g2:
                best_g2 = g2
                best_config = (x1, x2)
                
print(f"\nBest configuration found:")
print(f"  B1 = ({best_config[0]}, {best_config[0]**2})")
print(f"  B2 = ({best_config[1]}, {best_config[1]**2})")
print(f"  Achieves g(1)={8}, g(2)={best_g2}")
```

```python
# Cell: ENVELOPE DERIVATION - The Mathematical Breakthrough
# =========================================================
# KEY INSIGHT: The boundary we see is the ENVELOPE of the line families
# Lines go from each red point to parameter (a, a²) on parabola y=x²
#
# For each red point, we have a family of lines parameterized by 'a'.
# The envelope is found by eliminating 'a' from:
#   F(x,y,a) = 0  (line equation)
#   ∂F/∂a = 0     (envelope condition)

print("="*70)
print("DERIVING ENVELOPE EQUATION")
print("="*70)
print()

# Set up symbolic variables
var('x y a')

# Configuration
R1 = (0, 0)
R2 = (1, 0)
R3 = (0, 1)

print("Line families from red points to parabola y=x²:")
print(f"  R1 = {R1}")
print(f"  R2 = {R2}")
print(f"  R3 = {R3}")
print()

# === FAMILY 1: Lines from R1=(0,0) to (a, a²) ===
print("Family 1: Lines from R1=(0,0) to (a, a²)")
print("  Parametric: y = a*x")
# This family has degenerate envelope (the origin), skip

# === FAMILY 2: Lines from R2=(1,0) to (a, a²) ===
print("\nFamily 2: Lines from R2=(1,0) to (a, a²)")
print("  Line equation: (y - 0) / (x - 1) = (a² - 0) / (a - 1)")
print("  Simplified: y(a-1) = a²(x-1)")
print("  Rearranged: a²(x-1) - y*a + y = 0")

# Define F(x,y,a) for Family 2
F2 = a^2 * (x - 1) - y*a + y

# Envelope condition: ∂F/∂a = 0
dF2_da = diff(F2, a)
print(f"  ∂F/∂a = {dF2_da}")

# Solve for 'a' from ∂F/∂a = 0
print("\n  Solving ∂F/∂a = 0 for 'a':")
a_solutions_2 = solve(dF2_da == 0, a)
print(f"  a = {a_solutions_2}")

# Substitute back into F2=0 to get envelope
if a_solutions_2:
    a_sol = a_solutions_2[0].rhs()
    envelope_2 = F2.subs(a=a_sol)
    envelope_2_simplified = envelope_2.simplify_full()
    print(f"\n  Envelope equation (Family 2):")
    print(f"  {envelope_2_simplified} = 0")
    
# === FAMILY 3: Lines from R3=(0,1) to (a, a²) ===
print("\n" + "="*70)
print("Family 3: Lines from R3=(0,1) to (a, a²)")
print("  Line equation: (y - 1) / (x - 0) = (a² - 1) / (a - 0)")
print("  Simplified: y*a - a = x(a² - 1)")
print("  Rearranged: x*a² - y*a + (a - x) = 0")

# Define F(x,y,a) for Family 3
F3 = x*a^2 - y*a + a - x

# Envelope condition: ∂F/∂a = 0
dF3_da = diff(F3, a)
print(f"  ∂F/∂a = {dF3_da}")

# Solve for 'a' from ∂F/∂a = 0
print("\n  Solving ∂F/∂a = 0 for 'a':")
a_solutions_3 = solve(dF3_da == 0, a)
print(f"  a = {a_solutions_3}")

# Substitute back into F3=0 to get envelope
if a_solutions_3:
    a_sol = a_solutions_3[0].rhs()
    envelope_3 = F3.subs(a=a_sol)
    envelope_3_simplified = envelope_3.simplify_full()
    print(f"\n  Envelope equation (Family 3):")
    print(f"  {envelope_3_simplified} = 0")

print("\n" + "="*70)
print("ENVELOPE ANALYSIS COMPLETE")
print("Next: Plot envelopes and count lattice points")
print("="*70)
```

```python
# Cell: BOUNDARY ANALYSIS - Find the actual envelope from g(5) data
# ==================================================================
# The theoretical envelopes are complex. Let's analyze the ACTUAL boundary
# from the g(5) = 19,068 points we computed.

print("="*70)
print("ANALYZING BOUNDARY FROM g(5) DATA")
print("="*70)
print()

# We need to reload all_blues from the progression
# Let me recompute to get the final point set
print("Recomputing to get g(5) point set...")
reds = [R1, R2, R3]
all_blues_final = [B1, B2]

for day in range(1, 6):
    new_blues = compute_next_day(reds, all_blues_final)
    all_blues_final = all_blues_final + new_blues
    print(f"  Day {day}: {len(all_blues_final)} blues")

print(f"\nTotal g(5) = {len(all_blues_final):,} points")
print()

# Extract coordinates
x_coords = [float(b[0]) for b in all_blues_final]
y_coords = [float(b[1]) for b in all_blues_final]

# Find boundary points
x_min, x_max = min(x_coords), max(x_coords)
y_min, y_max = min(y_coords), max(y_coords)

print(f"Bounding box:")
print(f"  x ∈ [{x_min:.4f}, {x_max:.4f}]")
print(f"  y ∈ [{y_min:.4f}, {y_max:.4f}]")
print()

# Compute density in the region
area_approx = (x_max - x_min) * (y_max - y_min)
density = len(all_blues_final) / area_approx
print(f"Point density: {density:.2f} points per unit area")
print(f"Approximate region area: {area_approx:.2f}")
print()

# Find extreme points (candidates for envelope boundary)
print("Extreme points (boundary candidates):")
# Leftmost
leftmost_idx = x_coords.index(x_min)
print(f"  Leftmost: ({x_coords[leftmost_idx]:.6f}, {y_coords[leftmost_idx]:.6f})")
# Rightmost
rightmost_idx = x_coords.index(x_max)
print(f"  Rightmost: ({x_coords[rightmost_idx]:.6f}, {y_coords[rightmost_idx]:.6f})")
# Topmost
topmost_idx = y_coords.index(y_max)
print(f"  Topmost: ({x_coords[topmost_idx]:.6f}, {y_coords[topmost_idx]:.6f})")
# Bottommost
bottommost_idx = y_coords.index(y_min)
print(f"  Bottommost: ({x_coords[bottommost_idx]:.6f}, {y_coords[bottommost_idx]:.6f})")

print()
print("="*70)
```

```python
# Cell: SATURATION TEST - Attempt g(6) to verify hypothesis
# ==========================================================
# HYPOTHESIS: g(6) is taking long because we're near saturation
# If g(6) only adds a few hundred points, we've hit the boundary

import time
import signal

def timeout_handler(signum, frame):
    raise TimeoutError("Computation exceeded time limit")

print("="*70)
print("TESTING SATURATION HYPOTHESIS: Computing g(6)")
print("="*70)
print()
print("This will attempt g(5) → g(6) with 5-minute timeout")
print("If we're near saturation, very few new blues will be created")
print()

# Set 5-minute timeout
signal.signal(signal.SIGALRM, timeout_handler)
signal.alarm(300)  # 5 minutes = 300 seconds

try:
    start_time = time.time()
    
    # Compute g(6)
    print("Computing day 5 → 6...")
    new_blues_day6 = compute_next_day(reds, all_blues_final)
    
    elapsed = time.time() - start_time
    signal.alarm(0)  # Cancel alarm
    
    g6 = len(all_blues_final) + len(new_blues_day6)
    
    print(f"\n✓ COMPLETED in {elapsed:.1f}s")
    print(f"  g(5) = {len(all_blues_final):,}")
    print(f"  New blues on day 6: {len(new_blues_day6):,}")
    print(f"  g(6) = {g6:,}")
    print()
    
    # Analyze saturation
    growth_rate = len(new_blues_day6) / len(all_blues_final)
    print(f"Growth rate: {growth_rate:.6f} ({growth_rate*100:.4f}%)")
    print()
    
    if len(new_blues_day6) < 1000:
        print("*** SATURATION DETECTED ***")
        print(f"Only {len(new_blues_day6)} new points added!")
        print("The envelope boundary has been reached.")
        print()
        print("CONCLUSION: g(16) ≈ g(6) (or slightly higher)")
        print(f"ESTIMATE: g(16) ≈ {g6}")
    else:
        print(f"Still growing significantly ({len(new_blues_day6)} new points)")
        print("May need to compute further days...")
        
except TimeoutError:
    signal.alarm(0)
    print("\n⏱ TIMEOUT after 5 minutes")
    print("This confirms the computation is very slow.")
    print()
    print("INTERPRETATION:")
    print("  - g(6) requires checking ~1.6M line intersections")
    print("  - If only finding a few new points, most checks fail")
    print("  - This is consistent with saturation hypothesis")
    print()
    print("We need a different approach to estimate g(16)...")
    
print("="*70)
```
